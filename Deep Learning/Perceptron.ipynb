```python
import tensorflow as tf
```

```python
X = tf.constant([[0,0],[0,1],[1,0],[1,1]], dtype=tf.float32)
y = tf.constant([[0], [1], [1], [0]], dtype=tf.float32)
```

```python
perceptron = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', input_shape=(2,))
])

perceptron.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = perceptron.fit(X, y, epochs=100, callbacks=[tf.keras.callbacks.EarlyStopping(patience=3)])

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape(-1,784).astype('float32') / 255.0


mnist_perceptron = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid', input_shape=(784,))
])

mnist_perceptron.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
